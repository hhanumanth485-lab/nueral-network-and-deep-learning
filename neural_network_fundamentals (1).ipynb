```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 200

experience = np.random.randint(0, 10, n)
education = np.random.randint(1, 4, n)  # 1=UG, 2=PG, 3=PhD
skill_score = np.random.randint(1, 11, n)

# Salary formula (realistic logic + noise)
salary = (
    experience * 5000 +
    education * 8000 +
    skill_score * 3000 +
    np.random.normal(0, 5000, n)
)

data = pd.DataFrame({
    'experience': experience,
    'education': education,
    'skill_score': skill_score,
    'salary': salary
})

data.head()
df = pd.DataFrame(data)
df
```
